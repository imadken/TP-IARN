# Matrice de confusion from scratch

```python
def confusion_matrix_scratch(actual, predicted):
   
   matrix = np.zeros((10,10))
    
   for i in range(len(actual)):
    
      matrix[actual[i]-1][predicted[i]-1] += 1
      
   return np.array(matrix,dtype="int64")
```

Notons que l'étiquette 10 concerne le chiffre 0. L'étiquette a été utilisée ainsi afin de faciliter les calculs conséquents.

```python
conf_mat = confusion_matrix_scratch(Y,Y)
conf_mat
```

```python
def get_parameters_matrix(confusion_matrix):
    
  FP = confusion_matrix.sum(axis=0) - np.diag(confusion_matrix)  
  
  FN = confusion_matrix.sum(axis=1) - np.diag(confusion_matrix)
    
  TP = np.diag(confusion_matrix)

  TN = confusion_matrix.sum() - (FP + FN + TP)
    
  #print(confusion_matrix.sum())  
  return (TP,TN,FP,FN)
```

```python
TP,TN,FP,FN=get_parameters_matrix(matrice_confusion)
print(TP)
print(TN)
print(FP)
print(FN)
```

# Méthode 0: K- Plus Proches Voisins

```python

class Knn_scratch:
    #constructeur
    def __init__(self, k):
        self.k = k
         
    
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y
        
        
    def predict(self, X):
        
        predictions = []
        
        for x in X:
            
            # pour chaque données de l'ensemble de test "Y_test" n calcule la distances euclidean
            
            
            distances = np.sqrt(np.sum((self.X_train - x)**2, axis=1))
            
            # on recupere les indices des k-plus proches 
           
            nearest_neighbors = np.argsort(distances)[:self.k]
            
            # on récupére les labels des k plus proches 
            
            nearest_labels = self.y_train[nearest_neighbors]
            
            #on calcule la fréquences des classes dans les k-plus 
            counts = np.bincount(nearest_labels , minlength=len(np.unique(self.y_train)))
            
            predicted_label = np.argmax(counts)
            predictions.append(predicted_label)
            
        return np.array(predictions)
    
    
    def predict_proba(self, X):
       
        probas =[]
        print(probas)
        
        for x in X:
            
            # pour chaque données de l'ensemble de test "Y_test" n calcule la distances euclidean
            
            
            distances = np.sqrt(np.sum((self.X_train - x)**2, axis=1))
            
            # on recupere les indices des k-plus proches 
           
            nearest_neighbors = np.argsort(distances)[:self.k]
            
            # on récupére les labels des k plus proches 
            
            nearest_labels = self.y_train[nearest_neighbors]
            
            #on calcule la fréquences des classes dans les k-plus 
            counts = np.bincount(nearest_labels ,minlength=len(np.unique(self.y_train)))
            
            # on dévise les fréquence par k pour obtenir le vecteur de proba 
            
            proba = counts / self.k
            probas.append(proba)
        
        return np.array(probas)
  
```

Ici il faudra implémenter la méthode, puis la tester et vérifier les métriques en variant le nombre K
