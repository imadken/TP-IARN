# Descente du Gradient : Préparation des fonctions

1- Calcul du coût

Cette fonction servira à calculer le cout $J(\theta_0,\theta_1)$

Elle prendra l'ensemble de données d'apprentissage en entrée ainsi que les paramètres définis initialement

```python
#fonction por claculer les prediction de notre modele
def get_pred(X, theta): 
    pred=np.dot(X,theta)
    return pred
#fonction pour claculer la fonction cout
def computeCost(X, y, theta):
    pred=get_pred(X, theta)
    cost= (pred - y[:,None])**2
    return np.mean(cost)

    # idéalement, tracer le coût à chaque itération pour s'assurer que la descente du gradient est correcte
    
    # calculer le coût avec et sans vectorisation, 
    # comparer le temps de traitement
    
  
```

2- Fonction de la descente du gradient

Cette fonction mettra à jour les paramètres $\theta_0,\theta_1$ jusqu'à convergence: atteinte du nombre d'itérations max, ou dérivée assez petite.

```python
#une fonction pour calculer la derivé de la fonction cout
def get_deriv(X, y, theta):
    A=np.transpose(X)
    pred=get_pred(X, theta)-y[:,None]
    deriv=np.dot(A,pred)
    return deriv/(X.shape[0]*2)
```

```python
# pour la descente du gradient de plus de la valeur optimiser de theta on retourne aussi des données pour les exploité dans 
# des graphes 
# une liste de touts les thetas calculé
#une liste des cout calculé pour chaque theta
#on a aussi considérer l'option d'arreter la descent si on retouve que le cout ne change que d'une tres petit valeur entre iteration
def gradientDescent(X, y, theta, alpha, iterations,early_stopping=False):
    thetas=[]
    cost_history=[]
    epsilon = 1e-9
    for i in range(0,iterations):
        thetas.append(theta)
        cost_history.append(computeCost(X, y, theta))
        deriv=get_deriv(X, y, theta) 
        theta=theta - alpha*deriv
        if early_stopping and i > 0 and (cost_history[-2] - cost_history[-1]) < epsilon:
            break
    return theta,thetas,cost_history
    # garder aussi le cout à chaque itération 
    # pour afficher le coût en fonction de theta0 et theta1
```

Traçage du coût en fonction de theta0 et theta1

* fonction pour afficher le graph 3D des deux thetas et le cout 

```python


# on Definie une fonction pour calculé la tangente en précisant thetat et le cout et la derivé dans theta 

def tangent_plane(theta, cost, grad):

    
    norm = np.array([grad[0,0], grad[1,0], -1])
    norm = norm / np.linalg.norm(norm)
    
    # Defenir un plan de theta_0 and autour de  theta
    delta = 1
    t0, t1 = np.meshgrid(np.arange(theta[0]-delta, theta[0]+delta, 0.1),
                          np.arange(theta[1]-delta, theta[1]+delta, 0.1))
    
    # calculé les couts du plan puis retourner les valeur 
    t2 = cost + grad[0]*(t0-theta[0]) + grad[1]*(t1-theta[1])
    
    return t0, t1, t2

# Define the parameters theta, X, and y here

# definir un interval de  theta_0 and theta_1 et calculé les couts
theta_0, theta_1 = np.meshgrid(np.arange(theta[0] - 1, theta[0] + 1, 0.1), 
                               np.arange(theta[1] - 1, theta[1] + 1, 0.1))
cost_values = np.zeros((len(theta_0), len(theta_1)))

for i in range(len(theta_0)):
    for j in range(len(theta_1)):
        t = np.array([theta_0[i, j], theta_1[i, j]]).reshape(2, 1)
        cost_values[i, j] = computeCost(X, y, t)

# Calculé la derive et le cout por le theta optimal
cost = computeCost(X, y, theta)
grad = get_deriv(X, y, theta)

# Calculé les coordoné du plan tangente 
t0, t1, t2 = tangent_plane(theta, cost, grad)

# cree le graphe et les plot 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')


surf = ax.plot_surface(theta_0, theta_1, cost_values, cmap="viridis", linewidth=0.2)


ax.plot_surface(t0, t1, t2, alpha=0.2)
```

## Stochastic gradient descent (SGD)
qui est une variante de la descent du gradient comme le mini_batch mais vu qu l'ensemble d'apprentissage est trop petit on a trouver que cette variante est plus interesante  

le SGD est different de la descente du gradient normal vu que pour mettre les theta ajour il ne calcule pas la derivé sur le tout le dataset  mais seulement dansun point sélectioné aléatoirement et donc dans chaque episode en selection un certain nombre de point aléatoire pour le calcule de la dérivé

```python
#on redefinit légerment la fonction de calcule de dérivé pour calculé la derivé dans un points seulement
def get_derivs(X, y, theta):
    A=np.transpose(X)
    pred=get_pred(X, theta)-y
    deriv=np.dot(A,pred)
    return deriv/(X.shape[0]*2)
def stochasticGradientDescent(X, y, theta, alpha, iterations, early_stopping=False):
    thetas = []
    cost_history = []
    m = 5 # on calcule la dérivé dans 5 point pour chaque épisode 
    epsilon = 1e-9
    for i in range(iterations):
        cost_history.append(computeCost(X, y, theta))
        for j in range(m):
            rand_index = np.random.randint(0, m)
            X_j = X[rand_index,:].reshape(1, X.shape[1])
            y_j = y[rand_index].reshape(1, 1)
            deriv = get_derivs(X_j, y_j, theta)
            theta = theta - alpha * deriv
        thetas.append(theta)
        if early_stopping and i > 0 and (cost_history[-2] - cost_history[-1]) < epsilon:
            break
    return theta, thetas, cost_history
```

```python
theta = np.zeros((3, 1),dtype=np.longlong)
iterations = 1500
alpha = 1e-9
# Appel
theta,thetas,cost_history_s = gradientDescent(XMulti , yMulti, theta, alpha, iterations)
theta = np.zeros((3, 1),dtype=np.longlong)
theta,thetas,cost_history = stochasticGradientDescent(XMulti , yMulti, theta, alpha, iterations)
#graphe de comparaison 
plt.figure(figsize=(15, 7))
plt.plot(range(len(cost_history)), cost_history,label='sans modification lr=1e-9 avec descente du gradient stochastic')
plt.xlabel("Number of iterations")
plt.ylabel("Cost function")
plt.plot(range(len(cost_history_s)), cost_history_s,label='sans modification lr=1e-9 descente du gradient normale')
plt.xlabel("Number of iterations")
plt.ylabel("Cost function")
plt.legend()
plt.title("Gradient descent optimization of cost function for linear regression")
plt.show()
```
